```python
import pandas as pd
import matplotlib.pyplot as plt    
import seaborn as sn
import numpy as np
import tensorflow
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_dataset=ImageDataGenerator(rescale=1.2/255,shear_range=0.2,zoom_range=0.2,horizontal_flip=True)
```

```python

test_dataset=ImageDataGenerator(rescale=1.2/255,shear_range=0.2,zoom_range=0.2,horizontal_flip=True)
```

```python
train_set=train_dataset.flow_from_directory("../input/8-kinds-of-image-classification/data/train/",target_size=(200,200),batch_size=32,class_mode="categorical")
```

```python
test_set=test_dataset.flow_from_directory("../input/8-kinds-of-image-classification/data/test/",target_size=(200,200),
                                          batch_size=32,class_mode="categorical")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten,Dropout,Conv2D,MaxPooling2D
```

```python
Machine=Sequential()
Machine.add(Conv2D(128,(3,3),input_shape=(200,200,3),activation="relu"))
Machine.add(MaxPooling2D(pool_size=(2,2)))
Machine.add(Conv2D(64,(3,3),activation="relu"))
Machine.add(MaxPooling2D(pool_size=(2,2)))
Machine.add(Flatten())
Machine.add(Dense(units=200,activation="relu"))
Machine.add(Dense(units=8,activation="softmax"))
```

```python
Machine.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
Machine.fit_generator(train_set,steps_per_epoch=5,epochs=300,validation_data=test_set,validation_steps=2000,verbose=1)
```
