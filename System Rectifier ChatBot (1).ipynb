```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
import json
import pickle

import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
from keras.optimizers import SGD
import random
```

```python
each_words=[]
batches=[]
files=[]
ignore=["?","!"]
```

```python
import json

with open("D:/deva/Intent-26-03-2021", encoding="utf8") as file:
    #item = input("Enter item name: ")
    data = json.load(file)
    
```

```python
data
```

```python
for content in data["intents"]:
    for pattern in content["patterns"]:
        data1=nltk.word_tokenize(pattern)
        each_words.extend(data1)
        files.append((data1,content["tag"]))
        if content["tag"] not in batches:
            batches.append(content["tag"])
```

```python
each_words=[lemmatizer.lemmatize(data1.lower()) for data1 in each_words if data1 not in ignore]
each_words=sorted(set(each_words))
batches=sorted(set(batches))
print(len(files),"files")
print(len(batches),"batches",batches)
print(len(each_words),"Unique Words are Lemmatized",each_words)
pickle.dump(each_words,open("each_words.pkl","wb"))
pickle.dump(batches,open("batches.pkl","wb"))
```

```python
training_data=[]
box=[0]*len(batches)
for doc in files:
    bag=[]
    pattern_words=doc[0]
    pattern_words=[lemmatizer.lemmatize(word.lower()) for word in pattern_words]
    for data1 in each_words:
        bag.append(1) if data1 in pattern_words else bag.append(0)
    row=list(box)
    row[batches.index(doc[1])]=1
    training_data.append([bag,row])
random.shuffle(training_data)
training_data=np.array(training_data)
x=list(training_data[:,0])
y=list(training_data[:,1])
```

```python
machine=Sequential()
machine.add(Dense(444,input_shape=(len(x[0]),),activation="relu"))
machine.add(Dropout(0.50))
machine.add(Dense(244,activation="relu"))
machine.add(Dropout(0.50))
machine.add(Dense(len(y[0]),activation="sigmoid"))
gradiant=SGD(momentum=0.9,nesterov=True,lr=0.01,decay=1e-6)
machine.compile(loss="categorical_crossentropy",optimizer=gradiant,metrics=["accuracy"])
chat=machine.fit(np.array(x),np.array(y),epochs=250,batch_size=5,verbose=1)
machine.save("Machine Chatbot.h5",chat)
```
