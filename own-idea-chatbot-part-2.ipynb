```python
data=pd.read_csv("../input/chatbotdataset/FAQ_MachineLearningInterview_com.csv")
data.head()
```

```python
# load stop words
stop_word = stopwords.words('english')
```

```python
def clean(text):
    text=re.sub(r"#\w+"," ",text)
    text=re.sub(r"@\w+"," ",text)
    text=re.sub(r"\d+"," ",text)
    text=re.sub(r"http\S+"," ",text)
    text=re.sub("r<.*?>"," ",text)
    text=text.split()
    text=" ".join([word for word in text if not word in stop_word])
    return text    
```

```python
data["FAQ Questions from MachineLearningInterview.com"]=data["FAQ Questions from MachineLearningInterview.com"].apply(lambda x:clean(x))
data.head()
```

```python
import string
string.punctuation
re_punc = re.compile('[%s]' % re.escape(string.punctuation))
data["FAQ Questions from MachineLearningInterview.com"]=[re_punc.sub(" ",w) for w in data["FAQ Questions from MachineLearningInterview.com"]]
data["FAQ Questions from MachineLearningInterview.com"]
```

```python
data["FAQ Questions from MachineLearningInterview.com"]=[word.lower() for word in data["FAQ Questions from MachineLearningInterview.com"]]
data["FAQ Questions from MachineLearningInterview.com"]
```

```python
x=np.array(data["FAQ Questions from MachineLearningInterview.com"])
y=np.array(data["FAQ Answers from MachineLearningInterview.com"])
```

```python
tv=CountVectorizer()
print(tv)
```

```python
x=tv.fit_transform(x)
```

```python
x=x.toarray()
```

```python
from sklearn.naive_bayes import GaussianNB
machine=GaussianNB()
machine.fit(x,y)
```

```python
Prediction=machine.predict(x)
```
